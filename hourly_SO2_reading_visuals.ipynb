```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.pyplot import FuncFormatter
```

```python
%matplotlib inline
```

```python
abesan_df = pd.read_csv('QC3QC2QC1_AQ_5_min_abesan_040820_310721.csv')

abesan_df.NO2_ug_m3 = pd.to_numeric(abesan_df.NO2_ug_m3, errors='coerce').astype(float)
abesan_df.O3_ug_m3 = pd.to_numeric(abesan_df.O3_ug_m3, errors='coerce').astype(float)
abesan_df.NO_ug_m3 = pd.to_numeric(abesan_df.NO_ug_m3, errors='coerce').astype(float)
abesan_df.TVOC_ppb = pd.to_numeric(abesan_df.TVOC_ppb, errors='coerce').astype(float)
abesan_df.CO2_ppm = pd.to_numeric(abesan_df.CO2_ppm, errors='coerce').astype(float)

abesan_df["DateTime"] = pd.to_datetime(abesan_df["DateTime"])
```

```python
abesan_d = abesan_df.groupby(['site', pd.Grouper(key = 'DateTime', freq = '24H')]).agg('mean')

abesan = abesan_d.reset_index().sort_values(['DateTime'])
abesan['site'] = abesan['site'].str.upper()
print(abesan.info())
print(abesan['site'].unique())
abesan.tail()
```

```python
ik_df = pd.read_csv('QC3QC2QC1_AQ_5_min_ikorodu_040820_310721.csv')

ik_df.NO2_ug_m3 = pd.to_numeric(ik_df.NO2_ug_m3, errors='coerce').astype(float)
ik_df.O3_ug_m3 = pd.to_numeric(ik_df.O3_ug_m3, errors='coerce').astype(float)
ik_df.NO_ug_m3 = pd.to_numeric(ik_df.NO_ug_m3, errors='coerce').astype(float)
ik_df.SO2_ug_m3 = pd.to_numeric(ik_df.SO2_ug_m3, errors='coerce').astype(float)
ik_df.TVOC_ppb = pd.to_numeric(ik_df.TVOC_ppb, errors='coerce').astype(float)
ik_df.CO2_ppm = pd.to_numeric(ik_df.CO2_ppm, errors='coerce').astype(float)

ik_df["DateTime"] = pd.to_datetime(ik_df["DateTime"])
```
